```python
def simplex_projection(v, b=1):
    if b < 0:
        print('Radius of simplex is negative: %2.3f\n', b)
        return None
    v = np.minimum(v, 1e15)
    v = (v > 0) * v
    u = np.sort(v)[::-1]
    sv = np.cumsum(u)
    z = (u > (sv - b) / range(1, len(u) + 1))
    non_neg_indices = np.argwhere(z != 0)
    if len(non_neg_indices):
        rho = non_neg_indices[-1, -1]   
    else:
        rho = 0
    theta = np.maximum(0, (sv[rho] - b) / (rho + 1))
    return np.maximum(v - theta, 0)

def optimize_weights(X, **kwargs):
    x_0 =  np.ones(X.shape[1]) / float(X.shape[1])
    objective = lambda b: -np.sum(np.log(np.maximum(np.dot(X - 1, b) + 1, 0.0001)))
    cons = ({'type': 'eq', 'fun': lambda b: 1 - sum(b)},)
    while True:
        res = optimize.minimize(objective, x_0, bounds=[(0., 1.)] * len(x_0), 
                                constraints=cons, method='slsqp', **kwargs)
        EPS = 1e-7
        if (res.x < 0. - EPS).any() or (res.x > 1. + EPS).any():
            X = X + np.random.randn(1)[0] * 1e-5
            print('Optimal weights not found, trying again...')
            continue
        elif res.success:
            break
        else:
            if np.isnan(res.x).any():
                print('Solution does not exist, use zero weights.')
                res.x = np.zeros(X.shape[1])
            else:
                print('Converged, but not successfully.')
            break
    return res.x

def convert_prices(date0, date1, period, decay):
    start_date = datetime.strptime(date0, '%Y-%m-%d')
    end_date = datetime.strptime(date1, '%Y-%m-%d') + timedelta(days=decay, hours=23, minutes = 59, seconds=59)
    df = pd.read_csv('Data/Testing_data/Prices/data_' + period + '.csv')
    df['Gmt time'] = pd.to_datetime(df['Gmt time'])
    df.head(2)
    mask = (df['Gmt time'] > start_date) & (df['Gmt time'] <= end_date)
    df_fin = df.loc[mask]
    df_fin = df_fin.reset_index(drop=True)
    df_fin_date = np.array(df_fin['Gmt time'])
    df_fin_price = np.array(df_fin[['EURUSD', 'CADUSD', 'GBPUSD', 'JPYUSD']])
    df_fin_date
    close_prices = []
    if period == 'm':
        hours = 0
        minutes = 0
    elif period == 'h':
        hours = 0
        minutes = 59
    elif period == 'd':
        hours = 1
        minutes = 59
    for i in range(1, (end_date - start_date).days+2):
        day_analyzed_end = start_date + timedelta(hours=24)*i - timedelta(hours=hours, 
                                                                          minutes=minutes, 
                                                                          seconds=60)
        if day_analyzed_end >= datetime.strptime('2019-03-10', '%Y-%m-%d'):
            day_analyzed_end = day_analyzed_end - timedelta(hours=1)
        mask_close_pr = (df_fin['Gmt time'] == day_analyzed_end)
        close_pr = df_fin.loc[mask_close_pr]
        close_pr = close_pr.reset_index(drop=True)
        close_prices.append(np.array(close_pr[['EURUSD', 'CADUSD', 'GBPUSD', 'JPYUSD', 'Gmt time']]))
    prices_ok = []
    prices_ok_eur = []
    prices_ok_cad = []
    prices_ok_gbp = []
    prices_ok_jpy = []
    for i in range(len(df_fin_date)):
        price_ed = np.array([df_fin_price[i,:], df_fin_date[i]])
        price_eur = np.array([df_fin_price[i,0], df_fin_date[i]])
        price_cad = np.array([df_fin_price[i,1], df_fin_date[i]])
        price_gbp = np.array([df_fin_price[i,2], df_fin_date[i]])
        price_jpy = np.array([df_fin_price[i,3], df_fin_date[i]])
        prices_ok.append(price_ed)
        prices_ok_eur.append(price_eur)
        prices_ok_cad.append(price_cad)
        prices_ok_gbp.append(price_gbp)
        prices_ok_jpy.append(price_jpy)
    return prices_ok, close_prices, prices_ok_eur, prices_ok_cad, prices_ok_gbp, prices_ok_jpy

def convert_sentiment(curr_pair, date0, date1):
    start_date = datetime.strptime(date0, '%Y-%m-%d') 
    end_date = datetime.strptime(date1, '%Y-%m-%d') + timedelta(days=0, hours=23, minutes = 59, seconds=59)
    df = pd.read_csv('Data/Classification_data/' + curr_pair + '/CascadingTests/' + curr_pair + 'output.csv', header=None)
    df[5] = pd.to_datetime(df[5])
    mask = (df[5] > start_date) & (df[5] <= end_date)
    df_fin = df.loc[mask]
    df_fin = df_fin.reset_index(drop=True)
    df_fin = df_fin.iloc[::-1]
    sent_all_pd = []
    cum_sent = []
    start_days = []
    for i in range(1, (end_date - start_date).days+2):
        if i == 1:
            day_analyzed_start = start_date
        else:
            day_analyzed_start = day_analyzed_end + timedelta(seconds=1)
        day_analyzed_end = start_date + timedelta(hours=24)*i - timedelta(seconds=1)
        start_days.append(day_analyzed_start)
        sent_mask = (df_fin[5] >= day_analyzed_start) & (df_fin[5] <= day_analyzed_end)
        sent_day_data = df_fin.loc[sent_mask]
        sent_day_data = sent_day_data.reset_index(drop=True)
        sent_ok = []
        cum_sent_day = 0
        for i in range(sent_day_data.shape[0]):
            if sent_day_data[3][i][0] == '0':
                classif = 0
            elif sent_day_data[3][i][0] == '-':
                classif = -1
            elif sent_day_data[3][i][0] == '1':
                classif = 1
            sent_ed = np.array([classif, sent_day_data[5][i]])
            cum_sent_day += sent_ed[0] 
            sent_ok.append(sent_ed)
        cum_sent.append(cum_sent_day)
        for i in range(1, len(sent_ok)):
            sent_ok[i][0] += sent_ok[i-1][0]
        sent_all_pd.append(pd.DataFrame(sent_ok))
    sent_all_pd = pd.concat(sent_all_pd, axis=0)
    return cum_sent, sent_all_pd, start_days
```

```python
def stream_rbc(curr_pair):
#     df_all = pd.read_csv('/Users/marinalebedin/Desktop/GUI_report/RBC_news.csv')
    df_all = pd.DataFrame()
    dollar = '%D0%B4%D0%BE%D0%BB%D0%BB%D0%B0%D1%80' 
    euro = '%D0%B5%D0%B2%D1%80%D0%BE'
    pound = 'D1%84%D1%83%D0%BD%D1%82'
    ruble = '%D1%80%D1%83%D0%B1%D0%BB%D1%8C'
    selection = []
    selection.append(dollar)
    selection.append(euro)
    selection.append(pound)
    selection.append(ruble)
    for s in selection:
        urls = []
        titles = []
        dates = []
        categories = []
        heads_rel = []
        categ_rel = []
        dates_rel = []
        urls_rel = []
        dates_rel_clean = []
        url = "https://www.rbc.ru/search/?project=rbcnews&query=" + s
        page = urllib.request.urlopen(url)
        soup = BeautifulSoup(page)
        news = soup.find_all('a', class_='search-item__link')
        for i in news:
            urls.append(i.attrs['href'])
            titles.append(i.text.strip())
        for i in tnrange(len(urls), desc='Running analysis of news'):
            if urls[i][:5] == 'https':
                page = urllib.request.urlopen(urls[i])
                soup = BeautifulSoup(page)
                date = soup.find_all('span', class_='article__header__date')
                category = soup.find_all('a', class_='article__header__category')
                if category == []:
                    categories.append('NA')
                for d in date:
                    dates.append(d.attrs['content'])
                for c in category:
                    if c.attrs['href'][19:] == 'finances':
                        categories.append('Финансы')
                    elif c.attrs['href'][19:] == 'economics':
                        categories.append('Экономика')
                    else:
                        categories.append('Not relevant')  
            else:
                dates.append('NA')
                categories.append('NA')
        df = pd.DataFrame({'Url':urls, 'Heading':titles, 'Category':categories, 'Date':dates})
        for i in range(df.shape[0]):
            if df.Category[i] == 'Финансы' or df.Category[i] == 'Экономика':
                heads_rel.append(df.Heading[i].strip())
                categ_rel.append(df.Category[i])
                dates_rel.append(df.Date[i])
                urls_rel.append(df.Url[i].strip())        
        for i in dates_rel:
            i = i.replace('T', '   ')
            i = i.replace('+03:00', '')
            dates_rel_clean.append(i)
        df_rel = pd.DataFrame({'Heading':heads_rel, 'Category':categ_rel, 'Date and Time':dates_rel_clean, 'Url':urls_rel})
        df_all = df_all.append(df_rel)
    df_all = df_all.reset_index(drop=True)
    df_all.to_csv('RBC_news.csv')
    return df_all

def stream_bkc():
#     df = pd.read_csv('/Users/marinalebedin/Desktop/GUI_report/BKC_news.csv')
    url = 'https://bcs-express.ru/category/valyutnyy-rynok'
    page = urllib.request.urlopen(url)
    soup = BeautifulSoup(page)
    urls = []
    news = soup.find_all('a', class_='feed-item__head')
    for n in tnrange(len(news), desc='Looking for news'):
        urls.append('https://bcs-express.ru/' + news[n].attrs['href'])
    dates = []
    headers = []
    for u in tnrange(len(urls), desc='Analysing news'):
        page = urllib.request.urlopen(urls[u])
        soup = BeautifulSoup(page)
        date = soup.find_all('div', class_='article__info-time')
        for d in date:
            dates.append(d.text.strip())
        header = soup.find_all('h1', class_='article__header')
        for h in header:
            headers.append(h.text.strip())
    df = pd.DataFrame({'Heading':headers, 'Date and Time':dates, 'Url':urls})
    df.to_csv('BKC_news.csv', index=False)
    return df
```

```python
def sentiment_strategy(date0, date1, period, decay):
    def sentiment_strategy_run(pairname, date0, date1, period, decay):
        threshold = 50
        if pairname == 'EURUSD':
            column = 0
        elif pairname == 'CADUSD':
            column = 1
        elif pairname == 'GBPUSD':
            column = 2
        elif pairname == 'JPYUSD':
            column = 3
        prices_data = convert_prices(date0, date1, period, decay)
        sent_data = convert_sentiment(pairname, date0, date1)
        diff_days = len(prices_data[1]) - len(sent_data[2])
        #   print('Difference in days b/t pr and sent = {} \n'.format(diff_days))
        signals = []
        for i in range(len(sent_data[0])):
            if sent_data[0][i] > 0 and sent_data[0][i] >= threshold:
                signal = -1.0
            elif sent_data[0][i] >= 0 and sent_data[0][i] < threshold:
                signal = 0.0
            elif sent_data[0][i] < 0:
                signal = 1.0
            signals.append(signal)
        trading_data = pd.concat([pd.DataFrame(np.array(prices_data[1])[:,0][:,4], columns=['date']), 
                                  pd.DataFrame(sent_data[0], columns=['day_sent']), 
                                  pd.DataFrame(np.array(prices_data[1])[:,0][:,column], columns=[pairname]),
                                  pd.DataFrame(signals, columns=['trading_signal'])],
                                 axis=1)
        n_buys = (pd.DataFrame(signals)[0] == 1.0).sum()
        n_sells = (pd.DataFrame(signals)[0] == -1.0).sum()
        n_holds = (pd.DataFrame(signals)[0] == 0.0).sum()
        #   print('Number of long positions = {}'.format(n_buys))
        #   print('Number of short positions = {}'.format(n_sells))
        #   print('Number of "do nothing" positions = {}'.format(n_holds))
        #   print()
        delta = n_buys - n_sells
        #   if delta > 0:
        #       print('We need to SELL such a number of {} in order to close position: {}'.format(pairname[:3] ,abs(delta)))
        #   elif delta < 0:
        #       print('We need to BUY such a number of {} in order to close position: {}'.format(pairname[:3] ,abs(delta)))
        #   elif delta == 0:
        #       print('Do nothing. Our delta is : {}'.format(abs(delta)))
        trading_data_np = np.array(trading_data)
        daily_pl = []
        daily_pl_cum = []
        for i in range(trading_data_np.shape[0]-diff_days):
            pl = -trading_data_np[i][2] * trading_data_np[i][3]
            daily_pl.append(pl)
            daily_pl_cum.append(pl)
        for i in range(1, len(daily_pl_cum)):
            daily_pl_cum[i] += daily_pl_cum[i-1]

        init_inv = []
        init_inv.append(0)
        for i in range(1,len(daily_pl_cum)):
            if np.array(daily_pl_cum[:i]).min() >= 0:
                inv = 0
            else:
                inv = -np.array(daily_pl_cum[:i]).min()
            init_inv.append(inv)

        daily_pl_graph = []
        for i in range(len(init_inv)):
#             if init_inv[i] == 0:
#                 pl_gr = daily_pl[i]
#             else:
            pl_gr = daily_pl[i]
            daily_pl_graph.append(pl_gr)

        cum_pl_graph = daily_pl_graph.copy()
        for i in range(1, len(cum_pl_graph)):
            cum_pl_graph[i] += cum_pl_graph[i-1]

        pl_intra = np.array(daily_pl).sum()
        days_av_to_close = 2
        end = np.array(np.array(prices_data[1])[:,0][:,column])[len(np.array(prices_data[1])[:,0][:,column])-days_av_to_close:len(np.array(prices_data[1])[:,0][:,column])-1].mean()
        pl_final = delta*end + pl_intra

        for i in range(diff_days-1):
            cum_pl_graph.append(0)
        for i in range(diff_days - days_av_to_close):
            daily_pl_graph.append(0)
        for i in range(days_av_to_close):
            daily_pl_graph.append((delta*end)/days_av_to_close)
        cum_pl_graph.append(pl_final)
        trading_data = pd.concat([pd.DataFrame(np.array(prices_data[1])[:,0][:,4], columns=['date']), 
                                  pd.DataFrame(sent_data[0], columns=['day_sent']), 
                                  pd.DataFrame(np.array(prices_data[1])[:,0][:,column], columns=['close_price']),
                                  pd.DataFrame(signals, columns=['trading_signal']),
                                  pd.DataFrame(daily_pl, columns=['daily_pl']),
                                 pd.DataFrame(daily_pl_cum, columns=['cum_pl']),
                                    pd.DataFrame(init_inv, columns=['init_inv']),
                                 pd.DataFrame(daily_pl_graph, columns=['daily_return_graph']),
                                 pd.DataFrame(cum_pl_graph, columns=['cum_return_graph'])],
                                 axis=1)
        if trading_data['cum_pl'].min() >= 0:
            initial_investment = 1
        else:
            initial_investment = -trading_data['cum_pl'].min()
        cum_ret = (pl_final) / initial_investment
        return cum_ret, np.array(trading_data)[:-diff_days,0], np.array(trading_data)[-diff_days-1:,0], np.array(trading_data)[:-diff_days,7], np.array(trading_data)[-diff_days-1:,7], np.array(trading_data)[:-diff_days,8], np.array(trading_data)[-diff_days-1:,8]
    results = {}
    results['sentiment_EURUSD'] = sentiment_strategy_run('EURUSD', date0, date1, period, decay)
    results['sentiment_GBPUSD'] = sentiment_strategy_run('GBPUSD', date0, date1, period, decay)
    results['sentiment_CADUSD'] = sentiment_strategy_run('CADUSD', date0, date1, period, decay)
    results['sentiment_JPYUSD'] = sentiment_strategy_run('JPYUSD', date0, date1, period, decay)
    return results

#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#

def anticor_expert(data, weight_o, w):
    T, N = data.shape
    weight = weight_o.copy()
    if T < 2 * w:
        return weight
    LX_1 = np.log(data[T - 2 * w: T - w])
    LX_2 = np.log(data[T - w: T])
    mu_1 = np.mean(LX_1, axis=0)
    mu_2 = np.mean(LX_2, axis=0)
    M_cov = np.zeros((N, N))
    M_cor = np.zeros((N, N))
    n_LX_1 = LX_1 - np.repeat(mu_1.reshape(1, -1), w, axis=0)
    n_LX_2 = LX_2 - np.repeat(mu_2.reshape(1, -1), w, axis=0)
    Std_1 = np.diag(n_LX_1.T @ n_LX_1) / (w - 1)
    Std_2 = np.diag(n_LX_2.T @ n_LX_2) / (w - 1)
    Std_12 = Std_1.reshape(-1, 1) @ Std_2.reshape(1, -1)
    M_cov = n_LX_1.T @ n_LX_2 / (w - 1)
    M_cor[Std_12 == 0] = 0
    M_cor[Std_12 != 0] = M_cov[Std_12 != 0] / np.sqrt(Std_12[Std_12 != 0])
    claim = np.zeros((N, N))
    w_mu_2 = np.repeat(mu_2.reshape(-1, 1), N, axis=1)
    w_mu_1 = np.repeat(mu_2.reshape(1, -1), N, axis=0)
    s_12 = (w_mu_2 >= w_mu_1) & (M_cor > 0)
    claim[s_12] += M_cor[s_12]
    diag_M_cor = np.diag(M_cor)
    cor_1 = np.repeat(-diag_M_cor.reshape(-1, 1), N, axis=1)
    cor_2 = np.repeat(-diag_M_cor.reshape(1, -1), N, axis=0)
    cor_1 = np.maximum(0, cor_1)
    cor_2 = np.maximum(0, cor_2)
    claim[s_12] += cor_1[s_12] + cor_2[s_12]
    transfer = np.zeros((N, N))
    sum_claim = np.repeat(np.sum(claim, axis=1).reshape(-1, 1), N, axis=1)
    s_1 = np.abs(sum_claim) > 0
    w_weight_o = np.repeat(weight_o.reshape(-1, 1), N, axis=1)
    transfer[s_1] = w_weight_o[s_1] * claim[s_1] / sum_claim[s_1]
    transfer_ij = transfer.T - transfer
    weight -= np.sum(transfer_ij, axis=0)
    return weight

def anticor(data, tc, opts):
    def anticor_kernel(data, window, exp_ret, exp_w):
        for k in range(window - 1):
            exp_w[k] = anticor_expert(data, exp_w[k], k + 2)
        numerator = 0.
        denominator = 0.
        for k in range(window - 1):
            numerator += exp_ret[k] * exp_w[k]
            denominator += exp_ret[k]
        weight = numerator / denominator
        return weight / sum(weight), exp_w

    def anticor_run(data, tc, opts, window=30, verbose=False):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        day_weight = np.ones(m) / m
        day_weight_o = np.zeros(m)
        daily_portfolio = np.zeros((n, m))
        exp_ret = np.ones(window - 1)
        exp_w = np.ones((window - 1, m)) / m
        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')
        for t in tqdm.trange(n):
            if t >= 1:
                day_weight, exp_w = anticor_kernel(data[:t], window, exp_ret, exp_w)
            daily_portfolio[t] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = (data[t] @ day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
            for k in range(window - 1):
                exp_ret[k] *= data[t] @ exp_w[k]
            exp_ret = exp_ret / sum(exp_ret)
            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio
    results = {}
    results['Anticor'] = anticor_run(data, tc, opts)
    return results 
    
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
    
def anticor_anticor(data, tc, opts):
    def anticor_anticor_kernel(data, window, exp_ret, exp_w, data_day, exp_ret2, exp_w2):
        for k in range(window - 1):
            exp_w[k] = anticor_expert(data, exp_w[k], k + 2)
        for k in range(window - 1):
            exp_w2[k] = anticor_expert(data_day, exp_w2[k], k + 2)
        numerator = 0.
        denominator = 0.
        for k in range(window - 1):
            numerator += exp_ret2[k] * exp_w2[k]
            denominator += exp_ret2[k]
        weight1 = numerator / denominator
        weight = exp_w.T @ weight1
        return weight / sum(weight), exp_w, exp_w2

    def anticor_anticor_run(data, tc, opts, window=30, verbose=False):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        day_weight = np.ones(m) / m
        day_weight_o = np.zeros(m)
        daily_portfolio = np.zeros((n, m))
        exp_ret = np.ones(window - 1)
        exp_w = np.ones((window - 1, m)) / m
        exp_ret2 = np.ones(window - 1)
        exp_w2 = np.ones((window - 1, window - 1)) / (window - 1)
        data_day = np.zeros((n, window - 1))
        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')
        for t in tqdm.trange(n):
            if t >= 1:
                day_weight, exp_w, exp_w2 = anticor_anticor_kernel(data[:t], window, 
                                        exp_ret, exp_w, data_day[:t], exp_ret2, exp_w2)
            daily_portfolio[t] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = (data[t] @ day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            if cum_ret < 0:
                print("something went wrong: cum_ret < 0")
                return None

            day_weight_o = day_weight * data[t] / daily_ret[t]

            data_day[t] = data[t].reshape(1, -1) @ exp_w.T

            for k in range(window - 1):
                exp_w[k] = exp_w[k] * data[t] / data_day[t, k]
                exp_ret2[k] *= data_day[t] @ exp_w2[k]
                exp_w2[k] *= data_day[t] / (data_day[t] @ exp_w2[k])

            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))

    #     return cum_ret, cumprod_ret, daily_ret, daily_portfolio, exp_ret, exp_ret2
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio
    
    results = {}
    results['Anticor_Anticor'] = anticor_anticor_run(data, tc, opts)
    return results 

#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#

def pamr(data, tc, opts):
    def pamr_expert(x, weight_o, eta):
        weight = weight_o - eta * (x - np.mean(x))
        weight = simplex_projection(weight)
        if (weight < -0.00001).any() or (weight > 1.00001).any():
            str_print = 'pamr_expert: t=%d, sum(weight)=%f, returning uniform weights'
            print(str_print % (t, weight.sum()))
            return np.ones(len(weight_o)) / len(weight_o)
        return weight / sum(weight)

    def pamr_run(data, tc, opts, eps=0.5, C=500, variant=0, verbose=False):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        day_weight = np.ones(m) / m
        day_weight_o = np.zeros(m)
        daily_portfolio = np.zeros((n, m))
        eta = np.inf
        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')
        for t in tqdm.trange(n):
            if t >= 2:
                day_weight = pamr_expert(data[t - 1], day_weight, eta)
            daily_portfolio[t] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = np.dot(data[t], day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
            # update lagrange multiplier
            denom_part = data[t] - np.mean(data[t])
            if variant == 0:
                denominator = np.dot(denom_part, denom_part)
            elif variant == 1:
                denominator = np.dot(denom_part, denom_part)
            elif variant == 2:
                denominator = np.dot(denom_part, denom_part)
            else:
                print("Wrong variant parameter: must be 0, 1 or 2. Exiting.")
                return None
            if denominator != 0:
                eta = (daily_ret[t] - eps) / denominator
            eta = max(0, eta)
            eta = min(1e10, eta)

            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio
    
    results = {}
    results['PAMR'] = pamr_run(data, tc, opts)
    return results 

#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#

def rmr(data, tc, opts):
    def compute_L1_median(X, maxiter=200, tol=1e-10):
        y = np.median(X, 0)
        for i in range(maxiter):
            D = cdist(X, [y])
            nonzeros = (D != 0)[:, 0]
            Dinv = 1 / D[nonzeros]
            Dinvs = np.sum(Dinv)
            W = Dinv / Dinvs
            T = np.sum(W * X[nonzeros], 0)
            num_zeros = len(X) - np.sum(nonzeros)
            if num_zeros == 0:
                y1 = T
            elif num_zeros == len(X):
                return y / X[-1]
            else:
                R = (T - y) * Dinvs
                r = np.linalg.norm(R)
                rinv = 0 if r==0 else num_zeros/r
                y1 = max(0, 1-rinv)*T + min(1, rinv)*y
            if euclidean(y, y1) < tol:
                return y1 / X[-1]
            y = y1
        return y / X[-1]

    def rmr_next_weight(data_close, day_weight, w, epsilon):
        if data_close.shape[0] < w + 1:
            x_t1 = data_close[-1] / data_close[-2]
        else:
            x_t1 = compute_L1_median(data_close[-w:])
        denom = (np.linalg.norm(x_t1 - np.mean(x_t1)))**2 
        if denom == 0:
            alpha = 0
        else:
            alpha = min(0, (np.dot(x_t1, day_weight) - epsilon) / denom)
        alpha = min(100000, alpha)
        weight = day_weight - alpha * (x_t1 - np.mean(x_t1))
        weight = simplex_projection(weight)
        return weight / sum(weight)

    def rmr_run(data, tc, opts, epsilon=10, window=30, verbose=False):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        day_weight = np.ones(m) / m
        day_weight_o = np.zeros(m)
        daily_portfolio = np.zeros((n, m))
        data_phi = np.zeros(m)
        turno = 0
        data_close = np.ones((n + 1, m))
        for i in range(1, n + 1):
            data_close[i] = data_close[i - 1] * data[i - 1]
        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')
        for t in tqdm.trange(n):
            if t >= 1:
                day_weight = rmr_next_weight(data_close[:t + 1], 
                                               day_weight, window, epsilon)        
            daily_portfolio[t] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = (data[t] @ day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio
    
    results = {}
    results['RMR'] = rmr_run(data, tc, opts)
    return results 

#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#

def olmar(data, tc, opts):
    def olmar_next_weight(data, weight_o, epsilon, window):
        T, N = data.shape
        if T < window:
            data_phi = data[T - 1]
        else:
            data_phi = np.zeros(N)
            tmp_x = np.ones(N)
            for i in range(1, window + 1):
                data_phi = data_phi + 1. / tmp_x
                tmp_x = tmp_x * data[T - i]
            data_phi /= window
        ell = max([0, epsilon - data_phi @ weight_o])
        x_bar = np.mean(data_phi)    
        denom_part = data_phi - x_bar
        denominator = np.dot(denom_part, denom_part)
        if denominator != 0:
            lmbd = ell / denominator
        else:
            lmbd = 0
        weight = weight_o + lmbd * (data_phi - x_bar)
        weight = simplex_projection(weight)
        return weight / sum(weight)

    def olmar_run(data, tc, opts, epsilon=10, window=5, verbose=False):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        day_weight = np.ones(m) / m
        day_weight_o = np.zeros(m)
        daily_portfolio = np.zeros((n, m))
        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')
        day_weights = []
        for t in tqdm.trange(n):
            if t >= 1:
                day_weight = olmar_next_weight(data[:t], day_weight, epsilon, window)
            day_weights.append(day_weight.tolist())
            daily_portfolio[t] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = (data[t] @ day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))
        with open("./olmar_weights.json", "w") as f:
            json.dump(day_weights, f)
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio
    
    results = {}
    results['OLMAR'] = olmar_run(data, tc, opts)
    return results 

#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#

def ons(data, tc, opts):
    def find_projection_to_simplex(x, M):
        n = M.shape[0]
        P = matrix(2 * M)
        q = matrix(-2 * M @ x)
        G = matrix(-np.eye(n))
        h = matrix(np.zeros((n, 1)))
        A = matrix(np.ones((1, n)))
        b = matrix(1.)
        sol = solvers.qp(P, q, G, h, A, b)
        return np.squeeze(sol['x'])

    def ons_next_weight(x, last_weight, A, b, eta, beta, delta):
        N = len(x)
        grad = x / (x @ last_weight)
        grad = grad.reshape(-1, 1)
        hessian = -1 * (grad @ grad.T)
        A += (-1) * hessian
        b += (1 + 1 / beta) * grad
        A_inv = np.linalg.inv(A)
        weight = find_projection_to_simplex(delta * A_inv @ b, A)
        weight = (1 - eta) * weight + eta * np.ones(N) / N
        return weight / sum(weight)

    def ons_run(data, tc, opts, eta=0., beta=1., delta=0.125, verbose=False):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        day_weight = np.ones(m) / m
        day_weight_o = np.zeros(m)
        daily_portfolio = np.zeros((n, m))
        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')
        A = np.eye(m)
        b = np.zeros((m, 1))
        for t in tqdm.trange(n):
            if t >= 1:
                day_weight = ons_next_weight(data[t - 1], day_weight, A, b, 
                                                   eta, beta, delta)
            daily_portfolio[t, :] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = (data[t] @ day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio
    
    results = {}
    results['ONS'] = ons_run(data, tc, opts)
    return results 

#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#

def eg(data, tc, opts):
    def eg_next_weight(last_x, last_weight, eta):
        weight = last_weight * np.exp(eta * last_x / (last_x @ last_weight))
        return weight / sum(weight)

    def eg_run(data, tc, opts, eta=0.05, verbose=False):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        day_weight = np.ones(m) / m
        day_weight_o = np.zeros(m)
        daily_portfolio = np.zeros((n, m))
        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')
        for t in tqdm.trange(n):
            if t >= 1:
                day_weight = eg_next_weight(data[t - 1], day_weight, eta)
            daily_portfolio[t, :] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = (data[t] @ day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio
    
    results = {}
    results['EG'] = eg_run(data, tc, opts)
    return results 

#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#

def up(data, tc, opts):
    def up_next_weight(data, weight_o):
        del0 = 4e-3 # minimum coordinate
        delta = 5e-3 # spacing of grid
        M = 10 # number of samples
        S = 5 # number of steps in the random walk
        N = data.shape[1]
        r = np.ones(N) / N # Start each one at the uniform point
        b = np.ones(r.shape[0])
        allM = np.zeros((N, M)) # Take the average of m samples
        for m in range(M):
            b = r.copy()
            for i in range(S):
                bnew = b.copy()
                j = np.random.randint(N - 1)
                a = np.random.choice([-1, 1])
                bnew[j] = b[j] + (a * delta)
                bnew[N - 1] = b[N - 1] - (a * delta)
                if bnew[j] >= del0 and bnew[N - 1] >= del0:
                    muliplier_x = min(1, np.exp((b[N - 1] - (2 * del0)) / (N * delta)))
                    x = np.prod(data @ b) * muliplier_x
                    muliplier_y = min(1, np.exp((bnew[N - 1] - (2 * del0)) / (N * delta)))
                    y = np.prod(data @ bnew) * muliplier_y
                    pr = min(y / x, 1) # or pr = min(x / y, 1)
                    if np.random.rand() < pr:
                        b = bnew.copy()
            allM[:, m] = b
        weight = np.mean(allM, 1) # Taking the average of m samples.
        return weight / sum(weight)

    def up_run(data, tc, opts, verbose=False):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        day_weight = np.ones(m) / m
        day_weight_o = np.zeros(m)
        daily_portfolio = np.zeros((n, m))
        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')
        for t in tqdm.trange(n):
            if t >= 1:
                day_weight = up_next_weight(data[:t], day_weight)
            daily_portfolio[t, :] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = (data[t] @ day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio
    
    results = {}
    results['UP'] = up_run(data, tc, opts)
    return results 

#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#

def ucrp(data, tc, opts):
    def ucrp_run(data, tc, opts, day_weight=None, verbose=False):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        if day_weight is None:
            day_weight = np.ones(m) / m
        else:
            day_weight = np.array(day_weight)
        day_weight_o = np.zeros(m)
        daily_portfolio = np.zeros((n, m))
        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')
        for t in tqdm.trange(n):
            daily_portfolio[t] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = (data[t] @ day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio
    
    results = {}
    results['UCRP'] = ucrp_run(data, tc, opts)
    return results 

#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#

def best_stock(data, tc, opts):
    def best_stock_run(data, tc, opts, verbose=False):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        day_weight_o = np.zeros(m)
        daily_portfolio = np.zeros((n, m))
        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')
        best_stock_index = data.prod(axis=0).argmax()
        day_weight = np.zeros(m)
        day_weight[best_stock_index] = 1.
        for t in tqdm.trange(n):
            daily_portfolio[t, :] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = (data[t] @ day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio
    
    results = {}
    results['Best Stock'] = best_stock_run(data, tc, opts)
    return results 

#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#

def bcrp(data, tc, opts):
    def bcrp_run(data, tc, opts, verbose=False):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        day_weight = optimize_weights(data)
        day_weight_o = np.zeros(m)
        daily_portfolio = np.zeros((n, m))
        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')
        for t in tqdm.trange(n):
            daily_portfolio[t, :] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = (data[t] @ day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio
    
    results = {}
    results['BCRP'] = bcrp_run(data, tc, opts)
    return results

#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#

def ubah(data, tc, opts):
    def ubah_run(data, tc, opts, day_weight=None, verbose=False):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        if day_weight is None:
            day_weight = np.ones(m) / m
        else:
            day_weight = np.array(day_weight)
        day_weight_o = np.zeros(m)
        daily_portfolio = np.zeros((n, m))
        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')
        for t in tqdm.trange(n):
            if t >= 1:
                day_weight = day_weight_o / sum(day_weight_o)
            daily_portfolio[t, :] = day_weight
            tc_coef = (1 - tc  * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = (data[t] @ day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio
    
    results = {}
    results['UBAH'] = ubah_run(data, tc, opts)
    return results

#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#

def corn(data, tc, opts):
    def corn_expert(data, w, c):
        T, N = data.shape
        if T <= w:
            weight = np.ones(N) / N
            return weight
        if w == 0:
            histdata = data.copy()
        else:
            indices = []
            d2 = data[-w:].flatten()
            for i in range(w, T):
                d1 = data[i - w:i].flatten()
                if np.corrcoef(d1, d2)[0, 1] >= c:
                    indices.append(i)
            histdata = data[indices]
        if len(histdata) == 0:
            weight = np.ones(N) / N
        else:
            weight = optimize_weights(histdata)
        return weight / sum(weight)

    def corn_run(data, tc, opts, w=5, c=0.1, verbose=False):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        day_weight = np.ones(m) / m
        day_weight_o = np.zeros(m)
        daily_portfolio = np.zeros((n, m))
        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')
        for t in tqdm.trange(n):
            if t >= 1:
                day_weight = corn_expert(data[:t], w, c)
            daily_portfolio[t] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = data[t] @ day_weight * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio
    
    results = {}
    results['CORN'] = corn_run(data, tc, opts)
    return results

#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#

def arma(data, tc, opts):
    class Arima:
        def __init__(self, window=7, init_weights=None, 
                    epsilon=10**-0.5, learning_rate=1.75, method="ons",
                    lr_decrease_power=None):
            self.epsilon = epsilon
            if init_weights is None:
                self.weights = np.arange(1, window + 1) / 10
            else:
                self.weights = init_weights.copy()
            self.learning_rate = learning_rate
            self.window = window
            self.method = method
            self.A_trans = np.eye(self.window) / epsilon
            self.A = np.eye(self.window) * epsilon
            self.lr_decrease_power = lr_decrease_power

        def predict_next_price_arima_ogd(self, X):
            diff = self.weights @ X[:-1] - X[-1]
            grad = 2 * X[:-1] * diff
            self.weights -= 1 / self.learning_rate / self.lr_denom * grad
            next_price = self.weights @ X[-self.window:]
            return next_price

        def predict_next_price_arima_ons(self, X):
            diff = self.weights @ X[:-1] - X[-1]
            grad = 2 * X[:-1] * diff
            grad = grad.reshape(1, len(grad))
            self.A += grad.T @ grad
            B = (np.linalg.inv(self.A) @ grad.T).flatten()
            self.weights -= 1 / self.learning_rate / self.lr_denom * B
            next_price = self.weights @ X[-self.window:]
            return next_price

        def predict_next_price_arima_ons_simple(self, X):
            diff = self.weights @ X[:-1] - X[-1]
            grad = 2 * X[:-1] * diff
            grad = grad.reshape(1, len(grad))
            denom = 1 + grad @ self.A_trans @ grad.T
            numer = self.A_trans @ grad.T @ grad @ self.A_trans
            self.A_trans = self.A_trans -  numer / denom
            B = (grad @ self.A_trans).flatten()
            self.weights -= 1 / self.learning_rate / self.lr_denom * B
            next_price = self.weights @ X[-self.window:]
            return next_price

        def predict_next_price(self, X, lr_denom=1.):
            if self.lr_decrease_power is not None:
                self.lr_denom = lr_denom**self.lr_decrease_power
            else:
                self.lr_denom = 1
            if self.method == "ons":
                next_price = self.predict_next_price_arima_ons(X)
            elif self.method == "ons_simple":
                next_price = self.predict_next_price_arima_ons_simple(X)
            elif self.method == "ogd":
                next_price = self.predict_next_price_arima_ogd(X)
            else:
                print("Wrong method: ogd and ons are available")
                exit(1)
            return next_price

    def arma_next_weight(data_close, last_portfolio, reversion_threshold, arma, lr_denom=1.):
        assets_num = data_close.shape[1]
        next_prices = []
        for i in range(assets_num):
            X = data_close[-(arma.window + 1):, i]
            next_price = arma.predict_next_price(X, lr_denom)
            next_prices.append(next_price)
        next_prices = np.array(next_prices).flatten()
        next_price_relative = next_prices / data_close[-1]
        ell = max([0, reversion_threshold - next_price_relative @ last_portfolio])
        x_bar = np.mean(next_price_relative)    
        denom_part = next_price_relative - x_bar
        denominator = np.dot(denom_part, denom_part)
        if denominator != 0:
            lmbd = ell / denominator
        else:
            lmbd = 0
        weight = last_portfolio + lmbd * (next_price_relative - x_bar)
        weight = simplex_projection(weight)
        return weight / sum(weight)

    def arma_run(data, tc, opts, reversion_threshold=10, epsilon=10**-5.5,
                    learning_rate=1e3, window=7, verbose=False, method="ons",
                    init_weights_denom=1e4, lr_decrease_power=None):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        day_weight = np.ones(m) / m
        day_weight_o = np.zeros(m)
        daily_portfolio = np.zeros((n, m))
        data_phi = np.zeros(m)
        data_close = np.ones((n + 1, m))
        for i in range(1, n + 1):
            data_close[i] = data_close[i - 1] * data[i - 1]
        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')
        arma_weights = np.arange(1, window + 1) / init_weights_denom
        arma = Arima(window=window, init_weights=arma_weights,
                    epsilon=epsilon, learning_rate=learning_rate,
                    method=method, lr_decrease_power=lr_decrease_power)
        for t in tqdm.trange(n):
            if t >= window + 2:
                if lr_decrease_power is not None:
                    denom = t
                else:
                    denom = 1
                day_weight = arma_next_weight(data_close[:t + 1], day_weight, 
                                                reversion_threshold, arma, denom)
            daily_portfolio[t] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = (data[t] @ day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio
    
    arma_params = [0.5e3, 0.7, 7, 3, "ons", None]
    results = {}
    results['ARMA'] = arma_run(data, tc, opts, init_weights_denom=arma_params[0],
                              epsilon=arma_params[1], window=arma_params[2],
                              learning_rate=arma_params[3], method=arma_params[4],
                              lr_decrease_power=arma_params[5])
    return results

#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#

def arima(data, tc, opts):
    class Arima:
        def __init__(self, window=7, init_weights=None, epsilon=10**-0.5, learning_rate=1.75, method="ons",
                    lr_decrease_power=None):
            self.epsilon = epsilon
            if init_weights is None:
                self.weights = np.arange(1, window + 1) / 10
            else:
                self.weights = init_weights.copy()
            self.learning_rate = learning_rate
            self.window = window
            self.method = method
            self.A_trans = np.eye(self.window) / epsilon
            self.A = np.eye(self.window) * epsilon
            self.lr_decrease_power = lr_decrease_power

        def predict_next_price_arima_ogd(self, X):
            diff = self.weights @ X[:-1] - X[-1]
            grad = 2 * X[:-1] * diff
            self.weights -= 1 / self.learning_rate / self.lr_denom * grad
            next_price = self.weights @ X[-self.window:]
            return next_price

        def predict_next_price_arima_ons(self, X):
            diff = self.weights @ X[:-1] - X[-1]
            grad = 2 * X[:-1] * diff
            grad = grad.reshape(1, len(grad))
            self.A += grad.T @ grad
            B = (np.linalg.inv(self.A) @ grad.T).flatten()
            self.weights -= 1 / self.learning_rate / self.lr_denom * B
            next_price = self.weights @ X[-self.window:]
            return next_price

        def predict_next_price_arima_ons_simple(self, X):
            diff = self.weights @ X[:-1] - X[-1]
            grad = 2 * X[:-1] * diff
            grad = grad.reshape(1, len(grad))
            denom = 1 + grad @ self.A_trans @ grad.T
            numer = self.A_trans @ grad.T @ grad @ self.A_trans
            self.A_trans = self.A_trans -  numer / denom
            B = (grad @ self.A_trans).flatten()
            self.weights -= 1 / self.learning_rate / self.lr_denom * B
            next_price = self.weights @ X[-self.window:]
            return next_price

        def predict_next_price(self, X, lr_denom=1.):
            if self.lr_decrease_power is not None:
                self.lr_denom = lr_denom**self.lr_decrease_power
            else:
                self.lr_denom = 1
            if self.method == "ons":
                next_price = self.predict_next_price_arima_ons(X)
            elif self.method == "ons_simple":
                next_price = self.predict_next_price_arima_ons_simple(X)
            elif self.method == "ogd":
                next_price = self.predict_next_price_arima_ogd(X)
            else:
                print("Wrong method: ogd and ons are available")
                exit(1)
            return next_price

    def arima_next_weight(data_close, last_portfolio, reversion_threshold, arima, denom=1.):
        assets_num = data_close.shape[1]
        next_prices = []
        for i in range(assets_num):
            X = data_close[-(arima.window + 2):, i]
            diff = np.diff(X)
            next_price_diff = arima.predict_next_price(diff, denom)
            next_price = next_price_diff + X[-1]
            next_prices.append(next_price)
        next_prices = np.array(next_prices).flatten()
        next_price_relative = next_prices / data_close[-1]
        numerator = min([0, next_price_relative @ last_portfolio - reversion_threshold])
        x_bar = np.mean(next_price_relative)
        denom_part = next_price_relative - x_bar
        denominator = np.dot(denom_part, denom_part)
        if denominator != 0:
            lmbd = numerator / denominator
        else:
            lmbd = 0
        weight = last_portfolio - lmbd * (next_price_relative - x_bar)
        weight = simplex_projection(weight)
        return weight / sum(weight)

    def arima_run(data, tc, opts, reversion_threshold=10, epsilon=10**-0.5,
                    learning_rate=1.75, window=7, verbose=False, method="ons",
                    init_weights_denom=100, lr_decrease_power=None):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        day_weight = np.ones(m) / m
        day_weight_o = np.zeros(m)
        daily_portfolio = np.zeros((n, m))
        data_phi = np.zeros(m)
        data_close = np.ones((n + 1, m))
        for i in range(1, n + 1):
            data_close[i] = data_close[i - 1] * data[i - 1]
        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')
        arima_weights = np.arange(1, window + 1) / init_weights_denom
        arima = Arima(window=window, init_weights=arima_weights,
                    epsilon=epsilon, learning_rate=learning_rate, 
                    method=method, lr_decrease_power=lr_decrease_power)
        for t in tqdm.trange(n):
            if t >= window + 2:
                if lr_decrease_power is not None:
                    denom = t
                else:
                    denom = 1
                day_weight = arima_next_weight(data_close[:t + 1], day_weight, 
                                                reversion_threshold, arima, denom)
                if t > 15:
                    pass
            daily_portfolio[t] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = (data[t] @ day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))
        print("arima_weights before:", arima_weights)
        print("arima_weights after:", arima.weights)
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio
    
    arima_params = [10**5, 0.35, 7, 2.7e-4, "ons", 0.5]
    results = {}
    results['ARIMA'] = arima_run(data, tc, opts, init_weights_denom=arima_params[0],
                              epsilon=arima_params[1], window=arima_params[2],
                              learning_rate=arima_params[3], method=arima_params[4],
                              lr_decrease_power=arima_params[5])
    return results

#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
#-------------------------------------------------------------------------------------------------------------#
import numpy as np
import tqdm
from scipy.spatial.distance import cdist
from scipy.spatial.distance import euclidean

import json

def cfr_ogd(data, tc, opts):
    def compute_L1_median(X, maxiter=200, tol=1e-5):
        y = np.median(X, 0)
        for i in range(maxiter):
            D = cdist(X, [y])
            nonzeros = (D != 0)[:, 0]
            Dinv = 1 / D[nonzeros]
            Dinvs = np.sum(Dinv)
            W = Dinv / Dinvs
            T = np.sum(W * X[nonzeros], 0)
            num_zeros = len(X) - np.sum(nonzeros)
            if num_zeros == 0:
                y1 = T
            elif num_zeros == len(X):
                return y
            else:
                R = (T - y) * Dinvs
                r = np.linalg.norm(R)
                rinv = 0 if r==0 else num_zeros/r
                y1 = max(0, 1-rinv)*T + min(1, rinv)*y
            if euclidean(y, y1) < tol:
                return y1
            y = y1
        return y


    def simplex_projection(v, b=1):
        if b < 0:
            print('Radius of simplex is negative: %2.3f\n', b)
            return None
        v = np.minimum(v, 1e15)
        v = (v > 0) * v
        u = np.sort(v)[::-1]
        sv = np.cumsum(u)
        z = (u > (sv - b) / range(1, len(u) + 1))
        non_neg_indices = np.argwhere(z != 0)
        if len(non_neg_indices):
            rho = non_neg_indices[-1, -1]   
        else:
            rho = 0
        theta = np.maximum(0, (sv[rho] - b) / (rho + 1))
        return np.maximum(v - theta, 0)


    class Arima:
        def __init__(self, window=7, init_weights=None, 
                    epsilon=10**-0.5, learning_rate=1.75, method="ons",
                    lr_decrease_power=None):
            self.epsilon = epsilon
            if init_weights is None:
                self.weights = np.arange(1, window + 1) / 10
            else:
                self.weights = init_weights.copy()
            self.learning_rate = learning_rate
            self.window = window
            self.method = method
            self.A_trans = np.eye(self.window) / epsilon
            self.A = np.eye(self.window) * epsilon
            self.lr_decrease_power = lr_decrease_power

        def predict_next_price_arima_ogd(self, X):
            diff = self.weights @ X[:-1] - X[-1]
            grad = 2 * X[:-1] * diff
            self.weights -= 1 / self.learning_rate / self.lr_denom * grad
            next_price = self.weights @ X[-self.window:]
            return next_price

        def predict_next_price_arima_ons(self, X):
            diff = self.weights @ X[:-1] - X[-1]
            grad = 2 * X[:-1] * diff
            grad = grad.reshape(1, len(grad))
            self.A += grad.T @ grad
            B = (np.linalg.inv(self.A) @ grad.T).flatten()
            self.weights -= 1 / self.learning_rate / self.lr_denom * B
            next_price = self.weights @ X[-self.window:]
            return next_price

        def predict_next_price_arima_ons_simple(self, X):
            diff = self.weights @ X[:-1] - X[-1]
            grad = 2 * X[:-1] * diff
            grad = grad.reshape(1, len(grad))
            # Sherman-Morrison formula
            denom = 1 + grad @ self.A_trans @ grad.T
            numer = self.A_trans @ grad.T @ grad @ self.A_trans
            self.A_trans = self.A_trans -  numer / denom
            B = (grad @ self.A_trans).flatten()
            self.weights -= 1 / self.learning_rate / self.lr_denom * B
            next_price = self.weights @ X[-self.window:]
            return next_price

        def predict_next_price(self, X, lr_denom=1.):
            if self.lr_decrease_power is not None:
                self.lr_denom = lr_denom**self.lr_decrease_power
            else:
                self.lr_denom = 1
            if self.method == "ons":
                next_price = self.predict_next_price_arima_ons(X)
            elif self.method == "ons_simple":
                next_price = self.predict_next_price_arima_ons_simple(X)
            elif self.method == "ogd":
                next_price = self.predict_next_price_arima_ogd(X)
            else:
                print("Wrong method: ogd and ons are available")
                exit(1)
            return next_price


    class CfrOgd:
        def __init__(self, window, arma, arima, 
                    learning_rate=0.5, epsilon=10, estim_num=4):
            self.window = window
            self.arma = arma
            self.arima = arima
            self.estim_num = estim_num
            self.learning_rate = learning_rate
            self.epsilon = epsilon
            self.estimators_weights = np.ones(self.estim_num) / self.estim_num

            self.history_losses = []
            self.history_weighted_losses = []
            self.history_estimators_weights = []
            self.history_next_price_relatives = []
            self.history_predicted_prices = []

        def compute_loss(self, y_true, y_pred):
            return np.mean((y_true - y_pred)**2)

        def compute_loss_grad(self, y_true, y_pred, predicted_prices):
            loss = (y_pred - y_true)[:, np.newaxis] * predicted_prices
            return 2 * np.mean(loss, axis=0)

        def save_results(self, save_path):
            with open(save_path + "/losses", 'w') as f:
                json.dump(self.history_losses, f)
            with open(save_path + "/weighted_losses", 'w') as f:
                json.dump(self.history_weighted_losses, f)
            with open(save_path + "/estimators_weights", 'w') as f:
                json.dump(self.history_estimators_weights, f)
            with open(save_path + "/next_price_relatives", 'w') as f:
                json.dump(self.history_next_price_relatives, f)
            with open(save_path + "/predicted_prices", 'w') as f:
                json.dump(self.history_predicted_prices, f)

        def predict_prices(self, prices):
            instruments_number = prices.shape[1]
            predicted_prices = np.zeros((instruments_number, 
                                        self.estim_num))
            predicted_prices[:, 0] = compute_L1_median(prices[-self.window:])
            predicted_prices[:, 1] = prices[-self.window:].mean(axis=0)
            arma_next_prices = []
            for i in range(instruments_number):
#                 X = prices[-(self.arma.window + 1):, i]
#                 next_price = self.arma.predict_next_price(X)
#                 arma_next_prices.append(next_price)
                X = prices[-(self.arima.window + 2):, i]
                diff = np.diff(X)
                next_price_diff = self.arma.predict_next_price(diff)
                next_price = next_price_diff + X[-1]
                arma_next_prices.append(next_price)
            predicted_prices[:, 2] = np.array(arma_next_prices)
            if self.estim_num >= 4:
                arima_next_prices = []
                for i in range(instruments_number):
                    X = prices[-(self.arima.window + 2):, i]
                    diff = np.diff(X)
                    next_price_diff = self.arima.predict_next_price(diff)
                    next_price = next_price_diff + X[-1]
                    arima_next_prices.append(next_price)
                predicted_prices[:, 3] = np.array(arima_next_prices)
            return predicted_prices

        def cfr_ogd_next_price_relatives(self, prices, t):
            max_window = max([
                self.window, 
                self.arma.window + 1, 
                self.arima.window + 2
            ])
            prices_step = prices[-max_window - 1:-1]
            prices_current = prices[-1]
            predicted_prices = self.predict_prices(prices_step)
            weighted_prices = np.sum(predicted_prices * self.estimators_weights, axis=1)
            estimators_losses = []
            for i in range(self.estim_num):
                loss = self.compute_loss(prices_current, predicted_prices[:, i])
                estimators_losses.append(loss)

            weighted_loss = self.compute_loss(prices_current, weighted_prices)
            loss_grad = self.compute_loss_grad(prices_current, 
                                        weighted_prices, predicted_prices)
            self.estimators_weights -= self.learning_rate * loss_grad
            self.estimators_weights = simplex_projection(self.estimators_weights)
            self.estimators_weights /= sum(self.estimators_weights)        
            predicted_prices = self.predict_prices(prices[-max_window:])        
            next_price_relatives = np.sum(predicted_prices * \
                                        self.estimators_weights, axis=1) / prices[-1]
            self.history_losses.append(estimators_losses)
            self.history_weighted_losses.append(weighted_loss)
            self.history_estimators_weights.append(self.estimators_weights.tolist())
            self.history_predicted_prices.append(predicted_prices.tolist())
            self.history_next_price_relatives.append(next_price_relatives.tolist())
            return next_price_relatives

        def cfr_next_weight(self, prices, last_portfolio, t):
            if prices.shape[0] < self.window + 1:
                next_price_relative = prices[-1] / prices[-2]
            else:
                next_price_relative = self.cfr_ogd_next_price_relatives(prices, t)
            numerator = min([0, next_price_relative @ last_portfolio - self.epsilon])
            x_bar = np.mean(next_price_relative)
            denom_part = next_price_relative - x_bar
            denominator = np.dot(denom_part, denom_part)
            if denominator != 0:
                lmbd = numerator / denominator
            else:
                lmbd = 0
            weight = last_portfolio - lmbd * (next_price_relative - x_bar)
            weight = simplex_projection(weight)
            return weight / sum(weight)


    def cfr_ogd_run(data, tc, opts=None, epsilon=10, window=5,
                    learning_rate=0.05, estim_num=4, verbose=False,
                    arma_params=(1e4, 10**-5.5, 7, 1e3, "ogd", None),
                    arima_params=(100, 10**-0.5, 7, 1.75, "ogd", None)):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        day_weight = np.ones(m) / m
        day_weight_o = np.zeros(m)

        daily_portfolio = np.zeros((n, m))
        # close prices according to ratios
        data_close = np.ones((n + 1, m))
        for i in range(1, n + 1):
            data_close[i] = data_close[i - 1] * data[i - 1]

        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')

        arma_weights = np.arange(1, arma_params[2] + 1) / arma_params[0]
        arma = Arima(window=arma_params[2], init_weights=arma_weights,
                    epsilon=arma_params[1], learning_rate=arma_params[3],
                    method=arma_params[4], lr_decrease_power=arma_params[5])
        arima_weights = np.arange(1, arima_params[2] + 1) / arima_params[0]
        arima = Arima(window=arima_params[2], init_weights=arima_weights,
                    epsilon=arima_params[1], learning_rate=arima_params[3],
                    method=arima_params[4], lr_decrease_power=arima_params[5])
        cfr_agent = CfrOgd(window, arma, arima, learning_rate, 
                            estim_num=estim_num)
        day_weights = []
        #for t in tqdm.trange(n):
        for t in range(n):
            if t >= arima_params[2] + 2:
                if arma_params[5] is not None or arima_params[5] is not None:
                    denom = t
                else:
                    denom = 1
                day_weight = cfr_agent.cfr_next_weight(data_close[:t + 1], 
                                                        day_weight, denom)
            day_weights.append(day_weight.tolist())
            daily_portfolio[t] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = (data[t] @ day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))

        return cum_ret, cumprod_ret, daily_ret, daily_portfolio

#     arma_params = [0.5e3, 0.7, 7, 3, "ons", None]
#     arima_params = [10**5, 0.35, 7, 2.7e-4, "ons", 0.5]
    results = {}
#     results['CFR-OGD'] = cfr_ogd_run(data, tc, opts, epsilon=10, window=5, learning_rate=0.02,
#                             arma_params=arma_params, arima_params=arima_params)
    results['CFR-OGD'] = cfr_ogd_run(data, 
                tc=.0, opts=None, epsilon=10, window=5,
                learning_rate=0.05, estim_num=5, verbose=False,
                arma_params=(1e4, 10**-5.5, 7, 1e3, "ogd", None),
                arima_params=(100, 10**-0.5, 7, 1.75, "ogd", None))
    
    return results
```

```python
def cfr_ons_sa(data, data_time, period, tc, opts):
    def compute_L1_median(X, maxiter=200, tol=1e-5):
        y = np.median(X, 0)
        for i in range(maxiter):
            D = cdist(X, [y])
            nonzeros = (D != 0)[:, 0]
            Dinv = 1 / D[nonzeros]
            Dinvs = np.sum(Dinv)
            W = Dinv / Dinvs
            T = np.sum(W * X[nonzeros], 0)
            num_zeros = len(X) - np.sum(nonzeros)
            if num_zeros == 0:
                y1 = T
            elif num_zeros == len(X):
                return y
            else:
                R = (T - y) * Dinvs
                r = np.linalg.norm(R)
                rinv = 0 if r==0 else num_zeros/r
                y1 = max(0, 1-rinv)*T + min(1, rinv)*y
            if euclidean(y, y1) < tol:
                return y1
            y = y1
        return y

    def simplex_projection(v, b=1):
        if b < 0:
            print('Radius of simplex is negative: %2.3f\n', b)
            return None
        v = np.minimum(v, 1e15)
        v = (v > 0) * v
        u = np.sort(v)[::-1]
        sv = np.cumsum(u)
        z = (u > (sv - b) / range(1, len(u) + 1))
        non_neg_indices = np.argwhere(z != 0)
        if len(non_neg_indices):
            rho = non_neg_indices[-1, -1]   
        else:
            rho = 0
        theta = np.maximum(0, (sv[rho] - b) / (rho + 1))
        return np.maximum(v - theta, 0)


    class Arima:
        def __init__(self, window=7, init_weights=None, 
                    epsilon=10**-0.5, learning_rate=1.75, method="ons",
                    lr_decrease_power=None):
            self.epsilon = epsilon
            if init_weights is None:
                self.weights = np.arange(1, window + 1) / 10
            else:
                self.weights = init_weights.copy()
            self.learning_rate = learning_rate
            self.window = window
            self.method = method
            self.A_trans = np.eye(self.window) / epsilon
            self.A = np.eye(self.window) * epsilon
            self.lr_decrease_power = lr_decrease_power

        def predict_next_price_arima_ogd(self, X):
            diff = self.weights @ X[:-1] - X[-1]
            grad = 2 * X[:-1] * diff
            self.weights -= 1 / self.learning_rate / self.lr_denom * grad
            next_price = self.weights @ X[-self.window:]
            return next_price

        def predict_next_price_arima_ons(self, X):
            diff = self.weights @ X[:-1] - X[-1]
            grad = 2 * X[:-1] * diff
            grad = grad.reshape(1, len(grad))
            self.A += grad.T @ grad
            B = (np.linalg.inv(self.A) @ grad.T).flatten()
            self.weights -= 1 / self.learning_rate / self.lr_denom * B
            next_price = self.weights @ X[-self.window:]
            return next_price

        def predict_next_price_arima_ons_simple(self, X):
            diff = self.weights @ X[:-1] - X[-1]
            grad = 2 * X[:-1] * diff
            grad = grad.reshape(1, len(grad))
            # Sherman-Morrison formula
            denom = 1 + grad @ self.A_trans @ grad.T
            numer = self.A_trans @ grad.T @ grad @ self.A_trans
            self.A_trans = self.A_trans -  numer / denom
            B = (grad @ self.A_trans).flatten()
            self.weights -= 1 / self.learning_rate / self.lr_denom * B
            next_price = self.weights @ X[-self.window:]
            return next_price

        def predict_next_price(self, X, lr_denom=1.):
            if self.lr_decrease_power is not None:
                self.lr_denom = lr_denom**self.lr_decrease_power
            else:
                self.lr_denom = 1
            if self.method == "ons":
                next_price = self.predict_next_price_arima_ons(X)
            elif self.method == "ons_simple":
                next_price = self.predict_next_price_arima_ons_simple(X)
            elif self.method == "ogd":
                next_price = self.predict_next_price_arima_ogd(X)
            else:
                print("Wrong method: ogd and ons are available")
                exit(1)
            return next_price

    class CfrOgd:
        def __init__(self, window, arma, arima, 
                    learning_rate=0.5, epsilon=10, estim_num=4):
            self.window = window
            self.arma = arma
            self.arima = arima
            self.estim_num = estim_num
            self.learning_rate = learning_rate
            self.epsilon = epsilon
            self.estimators_weights = np.ones(self.estim_num) / self.estim_num

            self.history_losses = []
            self.history_weighted_losses = []
            self.history_estimators_weights = []
            self.history_next_price_relatives = []
            self.history_predicted_prices = []

        def compute_loss(self, y_true, y_pred):
            return np.mean((y_true - y_pred)**2)

        def compute_loss_grad(self, y_true, y_pred, predicted_prices):
            loss = (y_pred - y_true)[:, np.newaxis] * predicted_prices
            return 2 * np.mean(loss, axis=0)

        def save_results(self, save_path):
            with open(save_path + "/losses", 'w') as f:
                json.dump(self.history_losses, f)
            with open(save_path + "/weighted_losses", 'w') as f:
                json.dump(self.history_weighted_losses, f)
            with open(save_path + "/estimators_weights", 'w') as f:
                json.dump(self.history_estimators_weights, f)
            with open(save_path + "/next_price_relatives", 'w') as f:
                json.dump(self.history_next_price_relatives, f)
            with open(save_path + "/predicted_prices", 'w') as f:
                json.dump(self.history_predicted_prices, f)

        def predict_prices(self, prices, prices_time, ts_eur, ts_cad, ts_gbp, ts_jpy):
            instruments_number = prices.shape[1]
            predicted_prices = np.zeros((instruments_number, 
                                        self.estim_num))
            predicted_prices[:, 0] = compute_L1_median(prices[-self.window:])
            predicted_prices[:, 1] = prices[-self.window:].mean(axis=0)
            arma_next_prices = []
            for i in range(instruments_number):
#                 X = prices[-(self.arma.window + 1):, i]
#                 next_price = self.arma.predict_next_price(X)
#                 arma_next_prices.append(next_price)
                X = prices[-(self.arima.window + 2):, i]
                diff = np.diff(X)
                next_price_diff = self.arma.predict_next_price(diff)
                next_price = next_price_diff + X[-1]
                arma_next_prices.append(next_price)
                    
            predicted_prices[:, 2] = np.array(arma_next_prices)
            if self.estim_num >= 4:
                arima_next_prices = []
                for i in range(instruments_number):
                    X = prices[-(self.arima.window + 2):, i]
                    diff = np.diff(X)
                    next_price_diff = self.arima.predict_next_price(diff)
                    next_price = next_price_diff + X[-1]
                    arima_next_prices.append(next_price)
                predicted_prices[:, 3] = np.array(arima_next_prices)

             # sent an prices  
            sa_next_prices = []
            for i in range(instruments_number):
                if i == 0:
                    curr_pair = 'EURUSD'
                    sent_eur, next_price_sa = calc_sentiment(prices_time[-self.window:], ts_eur, curr_pair)
                elif i == 1:
                    curr_pair = 'CADUSD'
                    sent_cad, next_price_sa = calc_sentiment(prices_time[-self.window:], ts_cad, curr_pair)
                elif i == 2:
                    curr_pair = 'GBPUSD'
                    sent_gbp, next_price_sa = calc_sentiment(prices_time[-self.window:], ts_gbp, curr_pair)
                elif i == 3:
                    curr_pair = 'JPYUSD'
                    sent_jpy, next_price_sa = calc_sentiment(prices_time[-self.window:], ts_jpy, curr_pair)
                sa_next_prices.append(next_price_sa)

    #         predicted_prices[:, 4] = np.ones(4) # sent an price prediction
            predicted_prices[:, 4] = sa_next_prices

            print(predicted_prices)
            return predicted_prices, sent_eur, sent_cad, sent_gbp, sent_jpy

        def cfr_ogd_next_price_relatives(self, prices, prices_time, ts_eur, ts_cad, ts_gbp, ts_jpy, t):
            max_window = max([
                self.window, 
                self.arma.window + 1, 
                self.arima.window + 2
            ])
            prices_step = prices[-max_window - 1:-1]
            prices_step_time = prices_time[-max_window-1:-1] ######
            prices_current = prices[-1]
            predicted_prices, s_eur, s_cad, s_gbp, s_jpy = self.predict_prices(prices_step, 
                                                                               prices_step_time, 
                                                                               ts_eur,
                                                                               ts_cad,
                                                                               ts_gbp,
                                                                               ts_jpy) ######

            weighted_prices = np.sum(predicted_prices * self.estimators_weights, axis=1)
            estimators_losses = []
            for i in range(self.estim_num):
                loss = self.compute_loss(prices_current, predicted_prices[:, i])
                estimators_losses.append(loss)

            weighted_loss = self.compute_loss(prices_current, weighted_prices)
            loss_grad = self.compute_loss_grad(prices_current, 
                                        weighted_prices, predicted_prices)
            self.estimators_weights -= self.learning_rate * loss_grad
            self.estimators_weights = simplex_projection(self.estimators_weights)
            self.estimators_weights /= sum(self.estimators_weights)      

    #         predicted_prices = self.predict_prices(prices[-max_window:], prices_time[-max_window:])    ########   

            next_price_relatives = np.sum(predicted_prices * \
                                        self.estimators_weights, axis=1) / prices[-1]
            self.history_losses.append(estimators_losses)
            self.history_weighted_losses.append(weighted_loss)
            self.history_estimators_weights.append(self.estimators_weights.tolist())
            self.history_predicted_prices.append(predicted_prices.tolist())
            self.history_next_price_relatives.append(next_price_relatives.tolist())
            return next_price_relatives, s_eur, s_cad, s_gbp, s_jpy

        def cfr_next_weight(self, prices, prices_time, ts_eur, ts_cad, ts_gbp, ts_jpy, last_portfolio, t):
            if prices.shape[0] < self.window + 1:
                next_price_relative = prices[-1] / prices[-2]
            else:
                next_price_relative, s_eur, s_cad, s_gbp, s_jpy = self.cfr_ogd_next_price_relatives(prices, 
                                                                                                    prices_time, 
                                                                                                    ts_eur,
                                                                                                    ts_cad,
                                                                                                    ts_gbp,
                                                                                                    ts_jpy,
                                                                                                    t) 
            numerator = min([0, next_price_relative @ last_portfolio - self.epsilon])
            x_bar = np.mean(next_price_relative)
            denom_part = next_price_relative - x_bar
            denominator = np.dot(denom_part, denom_part)
            if denominator != 0:
                lmbd = numerator / denominator
            else:
                lmbd = 0
            weight = last_portfolio - lmbd * (next_price_relative - x_bar)
            weight = simplex_projection(weight)
            return weight / sum(weight), s_eur, s_cad, s_gbp, s_jpy


    def cfr_ogd_run(data, data_time, period, 
                    tc, opts=None, epsilon=10, window=5,
                    learning_rate=0.05, estim_num=4, verbose=False,
                    arma_params=(1e4, 10**-5.5, 7, 1e3, "ons", None),
                    arima_params=(100, 10**-0.5, 7, 1.75, "ons", None)):
        n, m = data.shape
        cum_ret = 1
        cumprod_ret = np.ones(n)
        daily_ret = np.ones(n)
        day_weight = np.ones(m) / m
        day_weight_o = np.zeros(m)

        daily_portfolio = np.zeros((n, m))
        # close prices according to ratios
        data_close = data
        data_close = np.ones((n + 1, m))
        data_close_time_ = np.ones((n + 1, m))
        for i in range(1, n + 1):
            data_close[i] = data_close[i - 1] * data[i - 1]
            data_close_time_[i] = data_close[i-1] * data_time[i-1, :-1]

        if period == 'h':
            first_date = data_time[0, -1] - timedelta(hours=1)
        data_close_time1 = np.column_stack((data_close_time_[:1], first_date))    
        data_close_time2 = np.column_stack((data_close_time_[1:], data_time[:,-1]))
        data_close_time = np.row_stack((data_close_time1, data_close_time2))

        if verbose:
            print('Parameters [tc: %f].\n' % tc)
            print('day\t Daily Return\t Total return\n')

        arma_weights = np.arange(1, arma_params[2] + 1) / arma_params[0]
        arma = Arima(window=arma_params[2], init_weights=arma_weights,
                    epsilon=arma_params[1], learning_rate=arma_params[3],
                    method=arma_params[4], lr_decrease_power=arma_params[5])
        arima_weights = np.arange(1, arima_params[2] + 1) / arima_params[0]
        arima = Arima(window=arima_params[2], init_weights=arima_weights,
                    epsilon=arima_params[1], learning_rate=arima_params[3],
                    method=arima_params[4], lr_decrease_power=arima_params[5])
        cfr_agent = CfrOgd(window, arma, arima, learning_rate, 
                            estim_num=estim_num)

        day_weights = []

        total_sent_eur = []
        total_sent_cad = []
        total_sent_gbp = []
        total_sent_jpy = []

        for t in tnrange(n):
            if t >= arima_params[2] + 2:
                if arma_params[5] is not None or arima_params[5] is not None:
                    denom = t
                else:
                    denom = 1
                day_weight, s_eur, s_cad, s_gbp, s_jpy = cfr_agent.cfr_next_weight(data_close[:t + 1], 
                                                                 data_close_time[:t+1],
                                                                 total_sent_eur,
                                                                 total_sent_cad,
                                                                 total_sent_gbp,
                                                                 total_sent_jpy,
                                                                 day_weight, 
                                                                 denom)
                total_sent_eur.append(s_eur)
                total_sent_cad.append(s_cad)
                total_sent_gbp.append(s_gbp)
                total_sent_jpy.append(s_jpy)

            day_weights.append(day_weight.tolist())
            daily_portfolio[t] = day_weight
            tc_coef = (1 - tc * sum(abs(day_weight - day_weight_o)))
            daily_ret[t] = (data[t] @ day_weight) * tc_coef
    #         daily_ret[t] = ((data_close[t]/data_close[t-1]) @ day_weight) * tc_coef
            cum_ret = cum_ret * daily_ret[t]
            cumprod_ret[t] = cum_ret
            day_weight_o = day_weight * data[t] / daily_ret[t]
    #         day_weight_o = day_weight * (data_close[t]/data_close[t-1]) / daily_ret[t]
            if verbose:
                if (t + 1) % opts['display_interval'] == 0:
                    print('%d\t%f\t%f\n' % (t + 1, daily_ret[t], cumprod_ret[t]))
        return cum_ret, cumprod_ret, daily_ret, daily_portfolio

#     arma_params = [0.5e3, 0.7, 7, 3, "ons", None]
#     arima_params = [10**5, 0.35, 7, 2.7e-4, "ons", 0.5]
    results = {}
#     results['CFR-OGD-SA'] = cfr_ogd_run(data, data_time, period, tc, opts, epsilon=10, window=5, 
#                                         learning_rate=0.02,
#                                         arma_params=arma_params, 
#                                         arima_params=arima_params)
    results['CFR-ONS-SA'] = cfr_ogd_run(data, data_time, period, 
                tc=.0, opts=None, epsilon=10, window=5,
                learning_rate=0.05, estim_num=5, verbose=False,
                arma_params=(1e4, 10**-5.5, 7, 1e3, "ons", None),
                arima_params=(100, 10**-0.5, 7, 1.75, "ons", None))
    
    return results

def calc_sentiment(data_time, prev_sent, curr_pair):
    time0 = data_time[0, -1]
    time1 = data_time[-1, -1]
    
    if curr_pair == 'EURUSD':
        column = 0
    elif curr_pair == 'CADUSD':
        column = 1
    elif curr_pair == 'GBPUSD':
        column = 2
    elif curr_pair == 'JPYUSD':
        column = 3
        
    df_fin_sa = pd.read_csv('Data/Classification_data/' + curr_pair + '/CascadingTests/' + curr_pair + 'output.csv', header=None)
    df_fin_sa[5] = pd.to_datetime(df_fin_sa[5])
    sent_mask = (df_fin_sa[5] >= time0) & (df_fin_sa[5] <= time1)
    sent_day_data = df_fin_sa.loc[sent_mask]
    sent_day_data = sent_day_data.reset_index(drop=True)

    cum_sent_day = 0
    for i in range(sent_day_data.shape[0]):
        if sent_day_data[3][i][0] == '0':
            classif = 0
        elif sent_day_data[3][i][0] == '-':
            classif = -1
        elif sent_day_data[3][i][0] == '1':
            classif = 1
        elif sent_day_data[3][i][0] == 'l':
            classif = 0 # 1
        sent_ed = np.array([classif, sent_day_data[5][i]])
        cum_sent_day += sent_ed[0] 
    
    mean_sent_period = np.array(prev_sent).mean()
    std_prices = np.array(data_time[:,column]).std()
    
    if cum_sent_day > mean_sent_period:
        pred_pr = data_time[-1, column] - std_prices #-
    elif cum_sent_day < mean_sent_period:
        pred_pr = data_time[-1, column] + std_prices #+
    else:
        pred_pr = data_time[-1, column]
        
    return cum_sent_day, pred_pr
```
